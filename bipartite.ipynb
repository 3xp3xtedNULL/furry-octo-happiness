```
import networkx as nx
import csv
import matplotlib.pyplot as plt
from networkx.algorithms import community
import community
```

```
# set the nodes
with open('/content/level1a - Sheet1.csv', 'r') as nodecsv:   
    # read the csv               
    nodereader = csv.reader(nodecsv) 
    time_mapping = []
    filler_mapping = []
    pause_mapping = []
    uhh_mapping = []
    nodes = []
    edges = []
    nodes.append('1')
    nodes.append('2')
    nodes.append('3')
    nodes.append('4')
    # get the nodes
    for n in nodereader:
        nodes.append(n[0])
        time_mapping.append(('1',n[0],{'weight' : n[5],'color':'black'}))
        time_mapping.append(('2',n[0],{'weight' : n[5],'color':'red'}))
        time_mapping.append(('3',n[0],{'weight' : n[5],'color':'blue'}))
        time_mapping.append(('4',n[0],{'weight' : n[5],'color':'green'}))
        filler_mapping.append(('1',n[0],{'weight' : n[4]}))
        filler_mapping.append(('2',n[0],{'weight' : n[4]}))
        filler_mapping.append(('3',n[0],{'weight' : n[4]}))
        filler_mapping.append(('4',n[0],{'weight' : n[4]}))
        pause_mapping.append(('1',n[0],{'weight' : n[3]}))
        pause_mapping.append(('2',n[0],{'weight' : n[3]}))
        pause_mapping.append(('3',n[0],{'weight' : n[3]}))
        pause_mapping.append(('4',n[0],{'weight' : n[3]}))
        uhh_mapping.append(('1',n[0],{'weight' : n[2]}))
        uhh_mapping.append(('2',n[0],{'weight' : n[2]}))
        uhh_mapping.append(('3',n[0],{'weight' : n[2]}))
        uhh_mapping.append(('4',n[0],{'weight' : n[2]}))
        edges.append(('1',n[0]))
        edges.append(('2',n[0]))
        edges.append(('3',n[0]))
        edges.append(('4',n[0]))

#nodes = nodes[2:]
#edges.pop(0)
print(nodes)
print(edges)
#mapping.pop(0)
print(time_mapping)
```

```
#Graph details for 
G0=nx.Graph()
G0.add_nodes_from(nodes)
#G.add_edges_from(edges)
G0.add_edges_from(time_mapping)

print("Nodes of graph: ")
print(G0.nodes())
print("Edges of graph: ")
print(G0.edges())
```

```
#Bipartite for Time Taken
plt.figure(3,figsize=(12,12))
pos=nx.circular_layout(G0) # pos = nx.nx_agraph.graphviz_layout(G)
nx.draw_networkx(G0,pos)
labels = nx.get_edge_attributes(G0,'weight')
nx.draw_networkx_edge_labels(G0,pos,edge_labels=labels,font_size = 10)
plt.savefig("Q1_Bipartite_TimeTaken.png")
```

```
print(nx.info(G0))
print('Degree centrality : ', max(nx.degree_centrality(G0).values()))
```
