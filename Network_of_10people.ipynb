```
import networkx as nx
import csv
import matplotlib.pyplot as plt
```

```
# set the nodes
with open('/content/Individual10people - Sheet1.csv', 'r') as nodecsv:   
    # read the csv               
    nodereader = csv.reader(nodecsv) 
    nodes = ['Filler','TimeTaken','Pause','Uh/uhm']
    edges = []
    # get the nodes
    for n in nodereader:
        if n[0] not in nodes:
          nodes.append(n[0])
        nodes.append((str(n[1])+n[0]))
        #mapping.append(('1',n[0],{'weight' : n[5]}))
        edges.append((n[0],(str(n[1])+n[0])))
        edges.append(((str(n[1])+n[0]),'Uh/uhm',{'weight' : n[2]}))
        edges.append(((str(n[1])+n[0]),'Pause',{'weight' : n[2]}))
        edges.append(((str(n[1])+n[0]),'TimeTaken',{'weight' : n[2]}))
        #edges.append(((str(n[1])+n[0]),'Pause',{'weight' : n[2]}))
        #edges.append(((str(n[1])+n[0]),'TimeTaken',{'weight' : n[2]}))
        edges.append(((str(n[1])+n[0]),'Filler',{'weight' : n[2]}))

print(nodes)
print(edges)
```

```
G=nx.Graph()
G.add_nodes_from(nodes)
#G.add_edges_from(edges)
G.add_edges_from(edges)

print("Nodes of graph: ")
print(G.nodes())
print("Edges of graph: ")
print(G.edges())
```

```
plt.figure(3,figsize=(20,20))
pos=nx.circular_layout(G) # pos = nx.nx_agraph.graphviz_layout(G)
nx.draw_networkx(G,pos)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels,font_size = 10)
plt.savefig("10people.png")
```

```
print(nx.info(G))
print('Degree centrality : ', max(nx.degree_centrality(G).values()))
```
