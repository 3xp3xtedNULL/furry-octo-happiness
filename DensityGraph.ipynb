```
import networkx as nx
import csv
import matplotlib.pyplot as plt
from networkx.algorithms import community
import community
```

```
# set the nodes
with open('/content/Level41a - Sheet1.csv', 'r') as nodecsv:   
    # read the csv               
    nodereader = csv.reader(nodecsv) 
    time_mapping = []
    filler_mapping = []
    pause_mapping = []
    uhh_mapping = []
    nodes = []
    nodes.append('4')
    # get the nodes
    for n in nodereader:
        nodes.append(n[0])
        time_mapping.append(('4',n[0],{'weight' : n[5]}))
        filler_mapping.append(('4',n[0],{'weight' : n[4]}))
        pause_mapping.append(('4',n[0],{'weight' : n[4]}))
        uhh_mapping.append(('4',n[0],{'weight' : n[2]}))
#nodes = nodes[2:]
#edges.pop(0)

#time_mapping.append(('4','K',{'weight': '0'}))
#time_mapping.append(('4','P',{'weight': '0'}))
nodes.append('E')
nodes.append('K')
nodes.append('P')
#time_mapping.append(('4','E',{'weight' : 0}))
#time_mapping.append(('4','K',{'weight' : 0}))
#time_mapping.append(('4','P',{'weight' : 0}))
print(nodes)

print(time_mapping)
print(filler_mapping)
print(pause_mapping)
print(uhh_mapping)

#mapping.pop(0)
```

```
G0=nx.Graph()
G0.add_nodes_from(nodes)
#G.add_edges_from(edges)
G0.add_edges_from(time_mapping)

print("Nodes of graph: ")
print(G0.nodes())
print("Edges of graph: ")
print(G0.edges())
```

```
plt.figure(3,figsize=(12,12))
pos=nx.spring_layout(G0) # pos = nx.nx_agraph.graphviz_layout(G)
nx.draw_networkx(G0,pos)
labels = nx.get_edge_attributes(G0,'weight')
nx.draw_networkx_edge_labels(G0,pos,edge_labels=labels,font_size = 10)
plt.savefig("Q4b_Density_TimeTaken.png")
```

```
print(nx.info(G0))
print('Degree centrality : ', max(nx.degree_centrality(G0).values()))
```
